```python
from spextra import Spextrum
import astropy.units as u

# Loading the S0 template from the Kinney-Calzetti library
sp = Spextrum("kc96/s0")  
# Scaling it to magnitude 22 AB in the sdss/g filter
sp2 = sp.scale_to_magnitude(amplitude=22 *u.ABmag, filter_name="g") 
# Add an emission line at 2000 A with flux 1e-16 erg/s/cm2 and 10A FWHM
sp3 = sp2.add_emi_lines(center=2000, flux=1e-16 * (u.erg/ (u.s * u.cm**2)) , fwhm=10 * u.AA)
# plotting
sp3.plot()
```
